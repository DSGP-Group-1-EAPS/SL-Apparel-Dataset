```python
# Define features and target variable
features = ['Encoded Reason', 'Encoded Status', 'Encoded Absenteeism Type', 'Encoded Shift', 'DaysWorked', 'DayOfWeek', 'Month']
target = 'TookLeaveNextMonth'


# Split data into training and testing sets
train = df[df['Date'] <= '2022-12-31']
test = df[df['Date'] > '2022-12-31']


# Create time series features
def create_time_series_features(df):
    """
    Create time series features based on 'Date' column.
    """
    df['hour'] = df['Date'].dt.hour
    df['dayofweek'] = df['Date'].dt.dayofweek
    df['quarter'] = df['Date'].dt.quarter
    df['month'] = df['Date'].dt.month
    df['year'] = df['Date'].dt.year
    df['dayofyear'] = df['Date'].dt.dayofyear
    df['dayofmonth'] = df['Date'].dt.day
    df['weekofyear'] = df['Date'].dt.isocalendar().week
    return df

# Apply function to create time series features
train = create_time_series_features(train)
test = create_time_series_features(test)

# Define features and target variables for training and testing
X_train = train[features]
y_train = train[target]

X_test = test[features]
y_test = test[target]
```

```python
# Create a time series with 'TookLeaveNextMonth' as the target variable
time_series = df.set_index('Date')[['TookLeaveNextMonth']]

# Stationarity Check
def check_stationarity(ts):
    # Perform Dickey-Fuller test
    result = adfuller(ts, autolag='AIC')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value}')

# Differencing to achieve stationarity if needed
time_series_diff = time_series.diff().dropna()
check_stationarity(time_series_diff)

# ARIMA Order Selection using ACF and PACF plots
plot_acf(time_series_diff)
plot_pacf(time_series_diff)
plt.show()

```

```python
# Define the values for p, d, and q
p = 1  # AR order
d = 1  # Differencing order
q = 1  # MA order

# Fit ARIMA model
order = (p, d, q)  # Specify appropriate values for p, d, and q
model = ARIMA(time_series, order=order)
results = model.fit()

# Make predictions
forecast_steps = len(df[df['Date'] > '2022-12-31'])
forecast = results.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=time_series.index[-1] + timedelta(days=1), periods=forecast_steps, freq='D')
forecast_df = pd.DataFrame({'prediction': forecast.predicted_mean.values}, index=forecast_index)

```

```python
# Convert probabilities to binary predictions (adjust the threshold as needed)
threshold = 0.5
test['binary_prediction'] = (forecast_df['prediction'] > threshold).astype(int)

# Ensure the 'binary_prediction' column is created in the test DataFrame before accessing it
accuracy = accuracy_score(test['TookLeaveNextMonth'], test['binary_prediction'])
print(f'Accuracy: {accuracy * 100:.2f}%')

# Employees predicted to be absent
predicted_absentees_ids = test.loc[test['binary_prediction'] == 1, 'Code']
print("Employees' IDs predicted to be absent next month:")
print(predicted_absentees_ids)


```
